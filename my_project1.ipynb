```python
import pyttsx3
import speech_recognition as sr
import webbrowser as wb
import smtplib
import imghdr
import os
from email.message import EmailMessage

speaker = pyttsx3.init()
speaker.say("hii this is your mail assitant")
speaker.say("do you wanna to send any mail")
speaker.runAndWait()

mic  =  sr.Microphone()
r = sr.Recognizer()

with mic as source:
    print("say :")
    audio = r.listen(source)
    text = r.recognize_google(audio)
    print(text)

    if "ok" or "yes" in text:
        print("what is ur mail_id")
        speaker.say("what is your mail id")
        speaker.runAndWait()  
        id=input("Enter mail id")


        EMAIL_ADDRESS=os.environ.get(id)

        print("what is ur password")
        speaker.say("what is your password")
        speaker.runAndWait()  
        password=input()

        EMAIL_PASSWORD=os.environ.get(password)
        msg=EmailMessage()

        msg['Subjecct']="just checking mail assistant"
        msg['From']=EMAIL_ADDRESS

        print("what is receiver mail id")
        speaker.say("what is receiver mail id")
        speaker.runAndWait() 
        rev=input("Enter R-mail id")

        msg['To']=rev
        
        print("do u wanna to attach image")
        speaker.say("do u wanna to attach image")
        speaker.runAndWait()
        print("say :")
        audio = r.listen(source)
        text = r.recognize_google(audio)
        print(text)
        
        if "yes" in text:
            
            print("enter your image path")
            speaker.say("enter your image path")
            speaker.runAndWait()  
            path_img=input("enter your image path")
        
            msg.set_content('image attached')
        
            print("what do wanna to write in content")
            speaker.say("what do wanna to write in content")
            speaker.runAndWait() 

            print("say :")
            audio = r.listen(source)
            text = r.recognize_google(audio)
            msg.set_content(text)

            with open(path_img,'rb')as f:
                file_data=f.read()
                file_type=imghdr.what(f.name)
                file_name=f.name
            msg.add_attachment(file_data,maintype='image',subtype='file_type',filename=file_name)
            with smtplib.SMTP_SSL('smtp.gmail.com',465)as smtp:
                smtp.login(id,password)
                smtp.send_message(msg)
            print("mail has sent") 
        else:
            
            print("what do wanna to write in content")
            speaker.say("what do wanna to write in content")
            speaker.runAndWait() 

            print("say :")
            audio = r.listen(source)
            text = r.recognize_google(audio)
            msg.set_content(text)
            print(text)
            with smtplib.SMTP_SSL('smtp.gmail.com',465)as smtp:
                smtp.login(id,password)
                smtp.send_message(msg)
            print("mail has sent") 

    else:
        print("please only give answer in yes or no ")
        speaker.say("please only give answer in yes or no")
        speaker.runAndWait()
   
```
